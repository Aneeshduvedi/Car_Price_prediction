```python
import numpy as np 
from sklearn.cluster import  DBSCAN
dbscan=DBSCAN(eps=0.06,min_samples=10)
dbscan.fit(data)
labels=dbscan.labels_
print(np.unique(labels))
print(np.unique(labels,return_counts=True))
```

```python
def out_detect(column):
    mean=column.mean()
    std=column.std()
    z_score=(column-mean)/std
    return column[(z_score>3) | (z_score<-3)]
out=out_detect(numerical["km_driven"])
f=numerical["km_driven"].isin(out)
numerical=numerical[~f]
category=category[~f]
#print(out)
out1=out_detect(numerical["max_power"])
f=numerical["max_power"].isin(out1)
numerical=numerical[~f]
category=category[~f]
#print(out1)
out2=out_detect(numerical["mileage"])
f=numerical["mileage"].isin(out2)
numerical=numerical[~f]
category=category[~f]
#print(out2)
```

```python
# one hot encoding on fuel and seller_type
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct=ColumnTransformer([("encode",OneHotEncoder(drop="first"),[6,7])],remainder="passthrough")
x=ct.fit_transform(x)
print(x)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
print(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#RANDOM FOREST REGRESSOR
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
rg=RandomForestRegressor(n_estimators=100)
rg.fit(x_train,y_train)
y_pred=rg.predict(x_test)
print(r2_score(y_test,y_pred))
```
